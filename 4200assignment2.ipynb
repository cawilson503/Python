```python
def main():
    n = int(input("Please enter a value for N:"))
    total = 0
    for i in range(0,n,2):
        total += ((1.0/(i+i+1))-(1.0)/(i+i+3))

    value = 4*total
    print(value)
```

```python
#Question 1 Palidromes

num = input("Enter a five digit number: ")
Q = (num[0])
X = (num[1])
W = (num[3])
Z = (num[4])

if Q == Z and X == W:
  print("The number is a palidrome.")

else:
  print("The number is not a palidrome.")
```

```python
#Question 2

for i in range(0,1000,2):
  total = 0
  denom = 1
  denom += i
  frac = (4/denom)
  if i % 2 == 0:
      total += frac
  else:
      total -= frac

  pi = 4-total
  print(pi)
```
