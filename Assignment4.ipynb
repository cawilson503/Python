```python
#2
import pandas as pd
import numpy as np
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
'age': [2.5, 3, 0.5, 'np.nan', 5, 2, 4.5, 'np.nan', 7, 3],
'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(data, index=labels)
print(df)
```

```python
print(df.info())
```

```python
print(df.loc['a':'c'])
```

```python
print(df[['animal', 'age']])
```

```python
print(df.iloc[[3,4,8], 0:2])
```

```python
print(df.loc[df['age'] == 'np.nan'])
```

```python
print(df['animal'].value_counts())
```

```python
print("Sum of visits:" ,df["visits"].sum())
```

```python
print(df.replace('snake', 'python'))
```

```python
df.replace('np.nan', np.nan, inplace=True)
df.dropna(inplace=True)
df['age'] = df['age'].astype(float)
df = df[df['animal'] == 'cat']
df['age'] = df['age'].astype(float)
df = df[df['age'] < 3]
print(df)
```

Question 4: Given a DataFrame of random numeric values: [5 Points]
df = pd.DataFrame(np.random.random(size=(5, 3))) # this is a 5x3 DataFrame
how do you subtract the row mean from each element in the row?
Hint: Use df.sub()

```python

import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.random(size=(5,3)))
print(df)
```

```python
print(df.mean(axis=1))
new_df = (df.sub(df.mean(axis=1),axis=0)) #you have to use axis=1 to get mean of rows, and axis=0 to subtract each element, syntax is
#(other, axis='columns', level=None, fill_value=None)
print(new_df)
```

```python
print("You have to use axis=1 to get mean of rows, and axis=0 to subtract it from each element when using df.sub(), otherwise the number of" '\n'
"rows and columns will be incorrect.")
```
